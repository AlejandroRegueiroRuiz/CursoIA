```python
import matplotlib.pyplot as plt
```

Utilizamos el teorema de balles en el caso de una mujer embarazada se hace varias pruebas hasta estar segura de que está embarazada

E = probabilidad de embarazo, NE = probabilidad de no estar embarazada. 

pP = probabilida de positivo, pN = probabilidad de negativo.

Creamos una función llamada balles que se encarga de recibir los parametros y devolver los parametros actualizados en el caso de salir positivo en el embarazo

```python
pE = 0.15
pNE = 0.85

pP_E = 0.9
pN_E = 0.1

pP_NE = 0.5
pN_NE = 0.5

def balles(pE, pNE, pP_E, pN_E, pP_NE, pN_NE):
    pPT = pE*pP_E + pNE*pP_NE

    pE_P = (pE*pP_E)/pPT
    pNE_P = (pNE*pP_NE)/pPT

    return pE_P, pNE_P

ListaPE = [pE]
ListaPNE = [pNE]
```

```python

for i in range(10):
    pE, pNE = balles(pE, pNE, pP_E, pN_E, pP_NE, pN_NE)
    ListaPE.append(pE)
    ListaPNE.append(pNE)
    print(pE, pNE)

# Creamos una grafica mostrando la evolucion de las probabilidades
plt.plot(ListaPE, label='P(E|P)')
plt.plot(ListaPNE, label='P(NE|P)')
plt.legend()
plt.show()
```
