```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

medicos_df = pd.DataFrame()
enfermeras_df = pd.DataFrame()

medicos_df = pd.read_csv('medicos.csv')
enfermeras_df = pd.read_csv('enfermeras.csv')

# Matplotlib: Distribution of ages of medicos and enfermeiros
plt.figure(figsize=(10, 6))
medicos_df['Fecha de Nacimiento'] = pd.to_datetime(medicos_df['Fecha de Nacimiento'], errors='coerce')
enfermeras_df['Fecha de Nacimiento'] = pd.to_datetime(enfermeras_df['Fecha de Nacimiento'], errors='coerce')
medicos_df['Edad'] = (pd.Timestamp.now() - medicos_df['Fecha de Nacimiento']).apply(lambda x: x.days // 365)
enfermeras_df['Edad'] = (pd.Timestamp.now() - enfermeras_df['Fecha de Nacimiento']).apply(lambda x: x.days // 365)
plt.hist(medicos_df['Edad'].dropna(), alpha=0.5, label='Medicos')
plt.hist(enfermeras_df['Edad'].dropna(), alpha=0.5, label='Enfermeras')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Distribución de edades de medicos y enfermeras')
plt.legend()
plt.show()

# Seaborn: Count plot of Sexo for medicos and enfermeiros
plt.figure(figsize=(10, 6))
sns.countplot(data=medicos_df, x='Sexo', label='Medicos', color='blue', alpha=0.5)
sns.countplot(data=enfermeras_df, x='Sexo', label='Enfermeras', color='orange', alpha=0.5)
plt.title('Distribución de Sexo en medicos y enfermeras')
plt.legend()
plt.show()

```
