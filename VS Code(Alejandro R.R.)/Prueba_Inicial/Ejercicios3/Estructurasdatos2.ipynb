Realizar un programa en Python que pida al usuario la dimensión y los elementos de dos listas de números enteros (pueden tener distinta dimensión):

- Una vez leídos los datos y almacenados en las listas, se generará una nueva lista que será la intersección de ambas, es decir, contendrá aquellos elementos que se encuentren en ambas listas.
- El vector intersección se mostrará por pantalla siempre que existan elementos comunes.

NOTA: Se comprobará que el usuario no introduzca elementos repetidos dentro del mismo vector.


#### Ejemplo de ejecución:
<pre>
Introduce el numero de elementos de la lista 1 (>0):  3
Introduce los números:
Dato 1: 2
Dato 2: 4
Dato 3: 1

Introduce el numero de elementos de la lista 2 (>0):  5
Introduce los números:
Dato 1: 3
Dato 2: 3
Error, elemento repetido. Dato 2: 2
Dato 3: 9
Dato 4: 0
Dato 5: 1

La intersección de las listas es: 2 1

-----------------------------------------------------------

Introduce el numero de elementos de la lista 1 (>0):  2
Introduce los números:
Dato 0: 3
Dato 1: -2

Introduce el numero de elementos de la lista 2 (>0):  3
Introduce los números:
Dato 0: 0
Dato 1: 4
Dato 2: 5

No hay elementos comunes
</pre>

```python
# Realiza tu solución aquí o en un archivo .py independiente

lista_1 = []
lista_2 = []

max = int(input("Introduce el numero de elementos de la lista 1 (>0): "))
while max <= 0:
    max = int(input("Introduce el numero de elementos de la lista 1 (>0): "))


while len(lista_1) < max:
    value = input(f"Dato {len(lista_1)}: ")
    while value in lista_1:
        print("Error, elemento repetido.")
        value = input(f"Dato {len(lista_1)}: ")
    lista_1.append(value)

max = int(input("Introduce el numero de elementos de la lista 2 (>0): "))
while max <= 0:
    max = int(input("Introduce el numero de elementos de la lista 2 (>0): "))

while len(lista_2) < max:
    value = input(f"Dato {len(lista_2)}: ")
    while value in lista_2:
        print("Error, elemento repetido.")
        value = input(f"Dato {len(lista_2)}: ")
    lista_2.append(value)


interseccion = set(lista_1) & set(lista_2)
if len(interseccion) == 0:
    print("No hay elementos comunes")
else:
    print(list(interseccion))



```

Escribir un programa que cree **un conjunto** simulando una lista de la compra. 

El programa debe preguntar el artículo y añadirlo al conjunto, mostrando un mensaje de aviso en caso de que el artículo ya existiese.



#### Ejemplo de ejecución:

<pre>
Introduce un artículo: pan
¿Quieres añadir artículos a la lista (Si/No)? Si
Introduce un artículo: agua
¿Quieres añadir artículos a la lista (Si/No)? Si
Introduce un artículo: leche
¿Quieres añadir artículos a la lista (Si/No)? Si
Introduce un artículo: café
¿Quieres añadir artículos a la lista (Si/No)? No
Artículos añadidos:
leche
agua
pan
café
Número de artículos:  4
</pre>

```python
# Realiza tu solución aquí o en un archivo .py independiente

lista_compra = []

art = input("Introduce un artículo: ")
lista_compra.append(art)

switch = input("¿Quieres añadir artículos a la lista (Si/No)? ")

while switch == "Si":
    art = input("Introduce un artículo: ")
    if art in lista_compra:
        print("El artículo ya está en la lista")
    else:
        lista_compra.append(art)
        
    switch = input("¿Quieres añadir artículos a la lista (Si/No)? ")

print("Artículos añadidos: ")
for artic in lista_compra:
    print(artic)

```

<h2 style="color: #007BC4;">Ejercicio 4</h2>
<hr style="border: 0.5px solid #007BC4;">

Realizar un programa en Python que permita gestionar las subscripciones de un usuario a diversas plataformas de entretenimiento, utilizando para ello **un diccionario**.

- El ejercicio parte de un diccionario que contiene el nombre y precio mensual de las subscripciones, formado por la clave (nombre de la plataforma) y el valor (precio mensual):


<pre>
        plataformas = {
        "Netflix": 11.99,
        "HBO": 8.99,       
        "Disney+": 7,
        "DAZN": 9.99,
        "Spotify": 9.99,
        "Apple Music": 9.99,
        "Amazon Prime": 3.99,
        }
</pre>


A continuación, deberá mostrar repetidamente un menú con las siguientes 6 opciones: 

1. Añadir plataforma (permite añadir una nueva plataforma al diccionario)
2. Eliminar plataforma (elimina una plataforma del diccionario)
3. Mostrar plataforma (muestra el precio de una plataforma a partir de su nombre)
4. Listar plataformas (muestra la lista de plataformas y sus precios)
5. Subscribirse n meses a una plataforma (muestra el precio total de la subscripción) 
6. Terminar (sale del programa)



#### Ejemplo de ejecución:

<img src="img/imagen1.png">

```python
# Realiza tu solución aquí o en un archivo .py independiente

plataformas = {
    "Netflix": 11.99,
    "HBO": 8.99,
    "Disney+": 7.00,
    "DAZN": 9.99,
    "Spotify": 9.99,
    "Apple Music": 9.99,
    "Amazon Prime": 3.99,
}

def añadir_plataforma():
    nombre = input("Introduce el nombre de la nueva plataforma: ")
    precio = float(input("Introduce el precio mensual de la plataforma: "))
    plataformas[nombre] = precio
    print(f"Plataforma {nombre} añadida correctamente.")

def eliminar_plataforma():
    nombre = input("Introduce el nombre de la plataforma a eliminar: ")
    if nombre in plataformas:
        del plataformas[nombre]
        print(f"Plataforma {nombre} eliminada correctamente.")
    else:
        print(f"No se encuentra la plataforma {nombre}.")

def mostrar_plataforma():
    nombre = input("Introduce el nombre de la plataforma: ")
    if nombre in plataformas:
        print(f"El precio mensual de {nombre} es {plataformas[nombre]:.2f}€.")
    else:
        print(f"No se encuentra la plataforma {nombre}.")

def listar_plataformas():
    print("Lista de plataformas y precios:")
    for nombre, precio in plataformas.items():
        print(f"{nombre}: {precio:.2f}€")

def subscribirse_meses():
    nombre = input("Introduce el nombre de la plataforma a la que te quieres suscribir: ")
    if nombre in plataformas:
        meses = int(input("Introduce el número de meses para la suscripción: "))
        total = plataformas[nombre] * meses
        print(f"El precio total por {meses} meses de {nombre} es {total:.2f}€.")
    else:
        print(f"No se encuentra la plataforma {nombre}.")

def menu():
    while True:
        print("\n----- Menú -----")
        print("1. Añadir plataforma")
        print("2. Eliminar plataforma")
        print("3. Mostrar plataforma")
        print("4. Listar plataformas")
        print("5. Subscribirse n meses a una plataforma")
        print("6. Terminar")
        
        opcion = input("Elige una opción (1-6): ")
        
        if opcion == "1":
            añadir_plataforma()
        elif opcion == "2":
            eliminar_plataforma()
        elif opcion == "3":
            mostrar_plataforma()
        elif opcion == "4":
            listar_plataformas()
        elif opcion == "5":
            subscribirse_meses()
        elif opcion == "6":
            print("¡Hasta luego!")
            break
        else:
            print("Opción no válida, por favor elige una opción entre 1 y 6.")

# Ejecutamos el menú
menu()








```
