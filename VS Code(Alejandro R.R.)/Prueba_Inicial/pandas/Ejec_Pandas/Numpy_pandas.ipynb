### 5. Genera una gráfica de barras donde se muestren el número propinas por día (con matplotlib o seaborn). 

```python
def graf_barras():
    tips_agrupados = tips_df.groupby('day').size().reset_index(name='Tips_dia')

    plt.bar(tips_agrupados['day'], tips_agrupados['Tips_dia'])
    plt.xlabel('Día de la semana')
    plt.ylabel('Número de propinas')

graf_barras()
plt.show()
```

### 6. Crea una gráfica **scatter** entre **total_bill** y **tip** utilizando matplotlib o seaborn. 

```python
total_bill = tips_df['total_bill']
tip = tips_df['tip']

plt.scatter(total_bill, tip)
plt.show()
```

### 7. Asigna un color a los puntos en base a la variable "sex" (muéstralo en la leyenda)

```python
def devolver_datos_bill_tip(sex):
    bill = tips_df[tips_df["sex"] == sex]["total_bill"]
    tips = tips_df[tips_df["sex"] == sex]["tip"]

    return bill,tips

def graf_scatter():
    total_bill_Male,tip_Male = devolver_datos_bill_tip("Male")
    plt.scatter(total_bill_Male, tip_Male, color="#ff3333", label="Male")

    total_bill_Female,tip_Female = devolver_datos_bill_tip("Female")
    plt.scatter(total_bill_Female, tip_Female, color="#80ff33", label="Female")
    plt.legend()
    return plt
graf_scatter()
plt.show()
```
