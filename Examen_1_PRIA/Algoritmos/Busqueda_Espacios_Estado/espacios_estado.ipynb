```python
import numpy as np
import random
import matplotlib.pyplot as plt

def flood_fill(matriz, x, y, nuevo_color):
    filas, columnas = matriz.shape
    color_original = matriz[x, y]
    if color_original == nuevo_color:
        return

    def dfs(x, y):
        if x < 0 or x >= filas or y < 0 or y >= columnas:
            return
        if matriz[x, y] != color_original:
            return
        matriz[x, y] = nuevo_color
        dfs(x + 1, y)
        dfs(x - 1, y)
        dfs(x, y + 1)
        dfs(x, y - 1)

    dfs(x, y)


# Generar matriz de ejemplo de forma aleatoria con valores 0, 127 y 255
valores = [0, 127, 255]
matriz_grayscale = np.random.choice(valores, (7, 5))

# Parámetros de ejemplo
fila = random.randint(0, 6)
columna = random.randint(0, 4)
nuevo_color = 128  # Verde

# Visualizar matriz original
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Mostrar matriz como círculos
for i in range(matriz_grayscale.shape[0]):
    for j in range(matriz_grayscale.shape[1]):
        if matriz_grayscale[i, j] == 0:
            ax[0].plot(j, i, 'ko', markersize=30)  # Negro
        elif matriz_grayscale[i, j] == 127:
            ax[0].plot(j, i, 'bo', markersize=30)  # Azul
        else:
            ax[0].plot(j, i, 'ro', markersize=30)  # Rojo

ax[0].set_title('Matriz Original')
ax[0].set_xlim(-1, matriz_grayscale.shape[1])
ax[0].set_ylim(-1, matriz_grayscale.shape[0])
ax[0].set_aspect('equal')

# Aplicar flood fill
flood_fill(matriz_grayscale, fila, columna, nuevo_color)

# Mostrar matriz modificada como círculos
for i in range(matriz_grayscale.shape[0]):
    for j in range(matriz_grayscale.shape[1]):
        if matriz_grayscale[i, j] == 0:
            ax[1].plot(j, i, 'ko', markersize=30)  # Negro
        elif matriz_grayscale[i, j] == 127:
            ax[1].plot(j, i, 'bo', markersize=30)  # Azul
        elif matriz_grayscale[i, j] == 128:
            ax[1].plot(j, i, 'go', markersize=30)  # Verde
        else:
            ax[1].plot(j, i, 'ro', markersize=30)  # Rojo

ax[1].set_title('Matriz con Flood Fill')
ax[1].set_xlim(-1, matriz_grayscale.shape[1])
ax[1].set_ylim(-1, matriz_grayscale.shape[0])
ax[1].set_aspect('equal')

plt.show()
```
