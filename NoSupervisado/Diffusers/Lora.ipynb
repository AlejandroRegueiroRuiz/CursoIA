```python
from diffusers import DiffusionPipeline
import torch
import os
store_path = os.getenv('STORE')
pipe_id = "stabilityai/stable-diffusion-xl-base-1.0"
#pipe_id = "latent-consistency/lcm-sdxl"
#pipe_id = "black-forest-labs/FLUX.1-schnell"
pipe = DiffusionPipeline.from_pretrained(pipe_id, torch_dtype=torch.float16, use_safetensors=True, variant="fp16", cache_dir=store_path)
pipe.to("cuda")
```

```python
# pipe.load_lora_weights("CiroN2022/toy-face", weight_name="toy_face_sdxl.safetensors", adapter_name="toy")

pipe.load_lora_weights("WiroAI/pokemon-flux-lora", weight_name="pokemon_flux_lora.safetensors", adapter_name="flux")
```

Asegúrate de usar el token *toy_face* en el prompt para poder realizar inferencia.

```python
prompt = "A dark, toxic creature with an aura of psychic energy. Its body is serpent-like, with glowing purple eyes and a venomous aura. It can manipulate its environment using psychic powers and releases poisonous gas from its mouth. The Pokémon is able to control minds with its psychic abilities, while its venom paralyzes its enemies. Its type is Psychic/Poison."

lora_scale = 0.9
image = pipe(
    prompt, num_inference_steps=30, cross_attention_kwargs={"scale": lora_scale}, generator=torch.manual_seed(0)
).images[0]
image
```
