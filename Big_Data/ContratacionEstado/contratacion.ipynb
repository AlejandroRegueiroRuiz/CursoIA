```python
# Crear un cursor para ejecutar las consultas
cursor = conexion.cursor()

# Cambiar el nombre de la tabla a un formato más amigable
table_name = "documentos"

# Crear la tabla si no existe
create_table_query = f"""
IF NOT EXISTS (SELECT * FROM sysobjects WHERE name='{table_name}' AND xtype='U')
CREATE TABLE {table_name} (
    Expediente NVARCHAR(MAX),
    [ID de publicación en TED] NVARCHAR(MAX),
    [Órgano de Contratación] NVARCHAR(MAX),
    [ID del Órgano de Contratación] NVARCHAR(MAX),
    [Estado de la Licitación] NVARCHAR(MAX),
    [Objeto del contrato] NVARCHAR(MAX),
    [Financiación UE] NVARCHAR(MAX),
    [Presupuesto base de licitación sin impuestos] NVARCHAR(MAX),
    [Valor estimado del contrato:] NVARCHAR(MAX),
    [Tipo de Contrato:] NVARCHAR(MAX),
    [Código CPV] NVARCHAR(MAX),
    [Lugar de Ejecución] NVARCHAR(MAX),
    [Sistema de contratación] NVARCHAR(MAX),
    [Procedimiento de contratación] NVARCHAR(MAX),
    [Tipo de tramitación] NVARCHAR(MAX),
    [Método de presentación de la oferta] NVARCHAR(MAX),
    [Fecha fin de presentación de oferta] NVARCHAR(MAX),
    [Resultado] NVARCHAR(MAX),
    [Adjudicatario] NVARCHAR(MAX),
    [Nº de Licitadores Presentados] NVARCHAR(MAX),
    [Importe de Adjudicación] NVARCHAR(MAX),
    [Lotes] NVARCHAR(MAX),
    [Fecha fin de presentación de solicitud] NVARCHAR(MAX)
)
"""
cursor.execute(create_table_query)

# Iterar sobre las filas del DataFrame
for index, row in df.iterrows():
    # Convertir todos los valores de la fila a string
    row = row.astype(str)


    # Crear la consulta de inserción
    query = f"""
    INSERT INTO {table_name} (
        Expediente,
        [ID de publicación en TED],
        [Órgano de Contratación],
        [ID del Órgano de Contratación],
        [Estado de la Licitación],
        [Objeto del contrato],
        [Financiación UE],
        [Presupuesto base de licitación sin impuestos],
        [Valor estimado del contrato:],
        [Tipo de Contrato:],
        [Código CPV],
        [Lugar de Ejecución],
        [Sistema de contratación],
        [Procedimiento de contratación],
        [Tipo de tramitación],
        [Método de presentación de la oferta],
        [Fecha fin de presentación de oferta],
        [Resultado],
        [Adjudicatario],
        [Nº de Licitadores Presentados],
        [Importe de Adjudicación],
        [Lotes],
        [Fecha fin de presentación de solicitud]
    ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    """
    
    # Ejecutar la consulta con los valores de la fila
    cursor.execute(query, row['Expediente'], row['ID de publicación en TED'], row['Órgano de Contratación'],
                   row['ID del Órgano de Contratación'], row['Estado de la Licitación'], row['Objeto del contrato'],
                   row['Financiación UE'], row['Presupuesto base de licitación sin impuestos'],
                   row['Valor estimado del contrato:'], row['Tipo de Contrato:'], row['Código CPV'],
                   row['Lugar de Ejecución'], row['Sistema de contratación'], row['Procedimiento de contratación'],
                   row['Tipo de tramitación'], row['Método de presentación de la oferta'],
                   row['Fecha fin de presentación de oferta'], row['Resultado'], row['Adjudicatario'],
                   row['Nº de Licitadores Presentados'], row['Importe de Adjudicación'], 
                   row.get('Lote / Descripción', 'No disponible'),
                   row['Fecha fin de presentación de solicitud'])

# Confirmar los cambios en la base de datos
conexion.commit()

# Cerrar el cursor
cursor.close()
```
