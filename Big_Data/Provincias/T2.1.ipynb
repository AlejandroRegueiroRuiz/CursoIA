1) Diferencias entre merge(), join() e concat. Pista: <https://realpython.com/pandas-merge-join-and-concat/>.

    <ul>Merge: Se utiliza para combinar DataFrames en base a una clave o múltiples claves, funciona de forma similar a un join de una sentencia SQL. Te permite especificar el tipo de join que quieres usar soportando el número de claves a mergear.</ul>
    <ul>Join: Funciona de forma similar al "merge" pero es de uso especifico para juntar DataFrames en función de sus índices, también puede elegir el tipo de join que quieres y si no especificar ninguno usará un "left join".</ul>
    <ul>Concat: Concatena diferentes DataFrames en base al axis introducido que pueden ser filas o columnas. Usado si quieres juntar DataFrames de forma vertical u horizontal. A diferencia de "merge" y "join" no requiere de una clave o índice en común.</ul>

```python
# A resolver, pista: <https://www.geeksforgeeks.org/selecting-rows-in-pandas-dataframe-based-on-conditions/>

def ContadorPublico(provincia):
    prov = df[(df['Provincia'] == provincia) & (df['TITULARIDADE'] == "Pública")]
    print(f"Hai {len(prov)} centros públicos en {provincia}")
    return prov

def ContadorCentros(SW):
    listProvincias = ["A Coruña", "Lugo", "Ourense", "Pontevedra"]
    listTipoCentro = ["CPI", "CEIP", "IES"]
    if SW == 0:
        for prov in listProvincias:
            c = df[(df['Provincia'] == prov) & (df['Tipo de centro'] == "CIFP")]
            print(f"Hai {len(c)} centros tipo CIFP en {prov}")
    elif SW == 1:
        for prov in listProvincias:
            for centro in listTipoCentro:
                c = df[(df['Provincia'] == prov) & (df['Tipo de centro'] == centro)]
                print(f"Hai {len(c)} centros tipo {centro} en {prov}")
            print("")
        


# Seleccionar centros que sexan de titularidade privada. Cantos hai?
df[df['TITULARIDADE'] == "Privada"]
# Seleccionar centros que teñan ensinanzas concertados. Cantos hai?
df[df['ENSINO_CONCERTADO'] == "Si"]
# Hai algún centro con ensinanzas concertadas dependente da Consellería?
concDep = df[(df['ENSINO_CONCERTADO'] == "Si") & (df['DEPENDENTE'] == "Si")]
if len(concDep) == 0:
    print("Non hai ningún centro concerdato que sexa dependente")
else:
    print("Si centro/s concerdato que sexa dependente")
# Seleccionar centros de titularidade pública de Coruña. Cantos hai?
publicCoruna_df = ContadorPublico("A Coruña")
publicCoruna_df
# Seleccionar centros de titularidade pública de Lugo. Cantos hai?
publicLugo_df = ContadorPublico("Lugo")
publicLugo_df
# Seleccionar centros de titularidade pública de Ourense. Cantos hai?
publicOurense_df = ContadorPublico("Ourense")
publicOurense_df
# Seleccionar centros de titularidade pública de Pontevedra. Cantos hai?
publicPontevedra_df = ContadorPublico("Pontevedra")
publicPontevedra_df
print("")
# Seleccionar centros de tipo: CIFP de cada unha das catro provincias. Cantos hai?
ContadorCentros(0)
print("")
# Seleccionar centros de tipo: CPI, CEIP e IES de cada unha das catro provincias. Cantos hai?
ContadorCentros(1)
print("")
# Seleccionar o número de centros por concello.
centrosConcello = df.groupby('Concello').size().reset_index(name='NumCentros')
centrosConcello
# Seleccionar o número de centros de cada tipo por concello.
centrosTipoConcello = df.groupby(['Concello', 'Tipo de centro']).size().reset_index(name='NumCentros')
centrosTipoConcello.head(20)

# Limpando datos con erros: Dime qué podes limpar e como nestes datos.
filas_con_nan = df[df.isnull().any(axis=1)]

# Para que no queden esos valores feos NaN, vamos a cambiar esos valores por un 0 indicando que no se a introducido ningún dato
filas_con_nan = filas_con_nan.fillna("0")

filas_con_nan
```

Garda o dataframe total resultante nos seguintes formatos (e logo volve abrilo nese formato e mostra a cabeceira):

**CSV**

```python
csv_temporal = os.path.join(tempfile.mktemp())

df.to_csv(csv_temporal, index = False)

CSV_df = pd.DataFrame
CSV_df = pd.read_csv(csv_temporal)

CSV_df.head()
```

**Excel**

Pasa a formato excel.

```python
excel_temporal = os.path.join(tempfile.mktemp(suffix='.xlsx'))

df.to_excel(excel_temporal, index = False)

Excel_df = pd.DataFrame
Excel_df = pd.read_excel(excel_temporal, sheet_name='Sheet1')

Excel_df.head()
```

**JSON**

Pasa a formato JSON.

(Coidado cos índices se xuntaches os dataframes)

```python
json_temporal = os.path.join(tempfile.mktemp(suffix='.json'))

df.to_json(json_temporal, index = False)

Json_df = pd.DataFrame
Json_df = pd.read_json(json_temporal)

Json_df.head()
```

**XML**

Pasa a formato XML:

```python
df.columns = df.columns.str.replace(' ', '_').str.replace('.', '').str.replace('ó', 'o')

XML_temporal = os.path.join(tempfile.mktemp(suffix='.xml'))

df.to_xml(XML_temporal, index = False)

XML_df = pd.DataFrame
XML_df = pd.read_xml(XML_temporal)

XML_df.head()
```

**PARQUET**

Pasa a formato parquet. Podes precisar as dependencias para **.parquet**: 

    `conda install pyarrow fastparquet´

```python
parquet_temporal = os.path.join(tempfile.mktemp(suffix='.parquet'))

df.to_parquet(parquet_temporal, engine='pyarrow')

parquet_df = pd.DataFrame
parquet_df = pd.read_parquet(parquet_temporal)

parquet_df.head()
```
